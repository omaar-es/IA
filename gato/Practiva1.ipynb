```python
import numpy as np

def N_Reynas():
    # Función para verificar si una reina puede colocarse en (fila, col) sin ser atacada
    def es_valido(tablero, fila, col, n):
        # Verifica que no haya otra reina en la misma columna
        for k in range(fila):
            if tablero[k][col] == 1:
                return False

        # Verifica la diagonal superior izquierda
        for k in range(1, n):
            if fila - k >= 0 and col - k >= 0 and tablero[fila - k][col - k] == 1:
                return False

        # Verifica la diagonal superior derecha
        for k in range(1, n):
            if fila - k >= 0 and col + k < n and tablero[fila - k][col + k] == 1:
                return False

        return True

    # Función principal que usa una pila para resolver el problema de manera iterativa
    def solve_iterativo(n):
        tablero = np.zeros((n, n), dtype=int)
        pila = []  # Pila para realizar la búsqueda en profundidad
        fila = 0
        col = 0

        while fila < n:
            colocado = False
            while col < n:
                if es_valido(tablero, fila, col, n):
                    tablero[fila][col] = 1
                    pila += [(fila, col)]  # Guarda la posición de la reina
                    fila += 1
                    col = 0
                    colocado = True
                    break
                col += 1

            if not colocado:
                if not pila:
                    return False, tablero  # No hay solución
                fila, col = pila[-1]  # Regresa a la última posición
                pila = pila[:-1]  # Elimina la última posición de la pila
                tablero[fila][col] = 0  # Elimina la reina de esa posición
                col += 1  # Intenta la siguiente columna

        return True, tablero

    # Solicita al usuario el número de reinas
    n = int(input("¿Cuántas reinas deseas colocar? "))

    # Llama a la función iterativa para resolver el problema
    resuelto, tablero_final = solve_iterativo(n)

    # Muestra el resultado
    if resuelto:
        print("Se ha encontrado una solución:")
        print(tablero_final)
    else:
        print("No se ha encontrado una solución.")

# Llamada a la función
N_Reynas()
```
